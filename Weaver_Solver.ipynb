```python
#Reads weaver word list
file_name = 'weaver_dict.txt'

with open(file_name, 'r') as file:
    weaver_list = file.read().splitlines()
```

```python
#Given word looks for all words that are within one letter
def possible_words(input_word):
    possible_word_list=[]
    for word in weaver_list:
        same=0
        i=0
        for letter in input_word:
            if word[i]==input_word[i] and len(input_word)==len(word):
                same=same+1
            i=i+1
        if same==3:
             possible_word_list.append(word)
    return possible_word_list

possible_words("kaon")
```

```python
#Given start and end word determines path
#creates queues of [word,[path]] and looks through all possibilities
#had to look up Breadth-First Search (BFS) algorithms
from collections import deque
def path(start_word, end_word):
    queue=deque([[start_word, [start_word]]
    ])
    visited=set([start_word])
    while queue:
        current_word, path = queue.popleft()
        neighbors=possible_words(current_word)

        if current_word==end_word:
            return path

        for word in neighbors:
            if word not in visited:
                visited.add(word)
                queue.append([word, path+[word]])
    return "No Path Found"

path("kaon", "iamb")




```
